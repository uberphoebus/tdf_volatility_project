```python
#   feature, label -> numpy type

def make_sequence_dataset(feature, label, window_size):
    feature_list = [] #  feature list
    label_list = []  #  label list
    
    for i in range(len(feature)-window_size):
        feature_list.append(feature[i:i+window_size])
        label_list.append(label[i+window_size])
    return np.array(feature_list), np.array(label_list) # 3 
```

```python
window_size = 20
X_train_1st, y_train_1st = make_sequence_dataset(X_train_1st_np, y_train_1st_np, window_size)
X_test_1st, y_test_1st = make_sequence_dataset(X_test_1st_np, y_test_1st_np, window_size)
print(X_train_1st.shape, y_train_1st.shape, X_test_1st.shape, y_test_1st.shape) # (batch size, time steps, input dims)
```

```python
# Initialising the RNN
model_117 = Sequential()

# Adding the first LSTM layer and some Dropout regularisation
model_117.add(LSTM(units = 10, activation = 'relu', return_sequences = True, input_shape = X_train_1st[0].shape))
model_117.add(Dropout(0.1))

# Adding a second LSTM layer and some Dropout regularisation
model_117.add(LSTM(units = 10, activation = 'relu', return_sequences = True))
model_117.add(Dropout(0.1))

# Adding a third LSTM layer and some Dropout regularisation
model_117.add(LSTM(units = 10, activation = 'relu', return_sequences = True))
model_117.add(Dropout(0.1))

# Adding a fourth LSTM layer and some Dropout regularisation
model_117.add(LSTM(units = 10, activation = 'relu'))
model_117.add(Dropout(0.1))

# Adding the output layer
model_117.add(Dense(units = 1))
model_117.summary()
# Compiling the RNN
# regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the RNN to the Training set
# regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import optimizers

# : https://kibua20.tistory.com/206 [ SW   ]
sgd = optimizers.SGD(learning_rate = 0.0001)
# sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model_117.compile(loss = 'mse', optimizer = 'adam')
early_stop = EarlyStopping(monitor = 'val_loss', patience = 10)
history = model_117.fit(X_train_1st, y_train_1st, validation_data = (X_test_1st, y_test_1st), epochs=100, 
          batch_size=15, callbacks = [early_stop])
```

```python
val_loss = history.history['val_loss']
print(val_loss[:5])
```

```python
train_loss = history.history['loss']
print(train_loss[:5])
```

```python
plt.plot(np.arange(len(val_loss)), val_loss, c="red", label='val')
plt.plot(np.arange(len(train_loss)), train_loss, c='blue', label ='train')
plt.legend()
plt.show()
```
